# Correlations

```python
# correlation matrix
display(data.corr())


# correlation plot
plt.figure(figsize = (16,6))
heatmap = sns.heatmap(data.corr(), vmin = -1, vmax = 1, annot = True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
plt.show()
```

```python
# correlation of categorical variables
from scipy.stats import chi2_contingency
import numpy as np

# get the columns
df = data[["WINNER", "INCUMBENT", "HOUSE", "SENATE"]]

# function for cramers_V
def cramers_V(var1,var2):
    crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
    stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
    obs = np.sum(crosstab) # Number of observations
    mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
    return (stat/(obs*mini))

# building matrix
rows= []

for var1 in df:
    col = []
    for var2 in df :
        cramers =cramers_V(df[var1], df[var2]) # Cramer's V test
        col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
    rows.append(col)
  
    
cramers_results = np.array(rows)
df_catcorr = pd.DataFrame(cramers_results, columns = df.columns, index =df.columns)

# display results
display(df_catcorr)

# correlation plot
plt.figure(figsize = (16,6))
heatmap = sns.heatmap(df_catcorr, vmin = -1, vmax = 1, annot = True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
plt.show()
```

**Observations:-**

- Following columns are highly correlated
    - DEM_VOTES, TOTALVOTES REP_VOTES. All three with POPULATION and TOTAL_ELIGIBLE_POPULATION
    - INCUMBENT and HOUSE are negatively correlated
    - POPULATION is completely correlated with TOTAL_ELIGIBLE_POPULATION
    - NETVOTES is correlated to DEM_VOTES
    - DEM_SPENDING is correlated with REP_SPENDING

**Steps taken:-**
    
- For regression analysis, we take ratio of dem_spending to rep_spending
- We include only one of HOUSE and INCUMBENT

# Deriving variables

## Generate derived variable of dem spending/ rep spending

```python
data["RATIO_SPENDING"] = data["DEM_SPENDING"]/data["REP_SPENDING"]

# view data to now groupby
```

## Log Transforming Variables

```python
data['ln_NETVOTES']=np.log(data['NETVOTES'])
data['ln_MEDIAN_INCOME']=np.log(data['MEDIAN_INCOME'])
data['ln_POPULATION']=np.log(data['POPULATION'])
data['ln_DEM_SPENDING']=np.log(data['DEM_SPENDING'])
data['ln_REP_SPENDING']=np.log(data['REP_SPENDING'])
data['ln_TOTAL_ELIGIBLE_POPULATION']=np.log(data['TOTAL_ELIGIBLE_POPULATION'])
data['ln_THIRD_PARTY_VOTES']=np.log(data['THIRD_PARTY_VOTES'])
data['ln_VEP_PCT_TURNOUT']=np.log(data['VEP_PCT_TURNOUT'])
data['ln_NET_VOTES_PCT']=np.log(data['NET_VOTES_PCT'])
```

# Datasets for regressions with state variable

## Data on state-by-state basis

```python
df_state = pd.get_dummies(data["STATE"])

# clean column names: case and spaces
df_state.columns = df_state.columns.str.upper()
df_state.columns = df_state.columns.str.strip().str.replace(' ', '_')

# concat
data_all = pd.concat([data, df_state], axis = 1)
data_all.drop(["STATE"], axis = 1, inplace = True)
data_all.drop(["WYOMING"], axis =1, inplace = True) # for avoiding the dummy variable trap: multicollinearity


with pd.option_context("display.max_columns", None):
    display(data_all)
```

## Data on yearly basis

```python
# View data for grouping
display(data)

# group
df_1 = data.groupby("YEAR").agg({"DEM_VOTES": "sum",
                                    "REP_VOTES": "sum",
                                    "TOTALVOTES": "sum",
                                    "MEDIAN_INCOME": "median",
                                    "POPULATION": "sum",
                                    "INCUMBENT": "mean",
                                    "HOUSE": "mean",
                                    "SENATE": "mean",
                                    "DEM_SPENDING": "mean",
                                    "REP_SPENDING": "mean",
                                    "VEP_PCT_TURNOUT":"mean",
                                    "TOTAL_ELIGIBLE_POPULATION": "sum",
                                    "RATIO_SPENDING": "mean",
                                    "NETVOTES": "sum",
                                     "NET_VOTES_PCT": "sum"
                                    })

df_2 = data.groupby("YEAR")[["WINNER"]].agg(pd.Series.mode)

df_group = pd.concat([df_1, df_2], axis = 1)

df_group.loc[2020, "WINNER"] = 1 # manually change winner for last year
display(df_group)
```

**Notes:-**

- Notice that NET_VOTES_PCT or NETVOTES column doesn't lead to a president winning an election, as seen in the final 4 rows. Therefore, if we want to do a regression analysis, we must use the entire dataset on state by state basis

**NOTE: We are not using data on yearly basis for our final analysis**

### Residual plots

#### For individual variables

```python
import statsmodels.api as sm

# residual plots for each variable
for col in ["MEDIAN_INCOME", "TOTAL_ELIGIBLE_POPULATION", "VEP_PCT_TURNOUT", "RATIO_SPENDING"]:
    fig = plt.figure(figsize=(12,8))
    fig = sm.graphics.plot_regress_exog(result1, col, fig=fig)
```

**Observations:-**
- The top right graph gives us residual vs plot
- We see that it might better to take natural log for TOTAL_ELIGIBLE_POPULATION (fanning in)

#### For predictions

```python
# get columns
columns = "MEDIAN_INCOME + TOTAL_ELIGIBLE_POPULATION + VEP_PCT_TURNOUT + RATIO_SPENDING +\
                 INCUMBENT + SENATE + " + states
columns = columns.replace(" ", "").split("+")
columns
```

```python
# predict
y_hat = result1.predict(data_all[columns])

# get residuals
residuals = data_all["NET_VOTES_PCT"] - y_hat

# plot
plt.scatter(residuals,y_hat)

plt.show()
```

**Observations:-**

- There don't seem to be any heterscedasticity in our residual plot for y_hat. Therefore, we don't need to log transform the same
