```python
import pandas as pd
import pickle

pd.options.display.float_format = '{:,.2f}'.format
```

```python
#historical population dates

df = pd.read_excel('popest-annual-historical.xls',  sheet_name= 'States')

df = df.drop(columns = 'Fips') # dropping unneeded column


df.set_index('Area Name', inplace = True)

df.drop(index=['United States', 'New England', 'District of Columbia','Mideast', 'Great Lakes', 'Plains', 'Southeast', 'Southwest', 'Rocky Mountain', 'Far West'], inplace = True)
# dropping uneeded rows

df = df[[1984, 1988, 1992, 1996, 2000, 2004, 2008]]

df.head(3)
```

```python
# population for last decade

df2 = pd.read_excel('popest-annual.xls',  sheet_name= 'States')

df2 = df2.drop(columns = ['Fips', 'Estimates Base (4/1/2010)']) # dropping unneeded columns

df2.set_index('Area', inplace = True)

df2 = df2[[2012, 2016]]

df2.head(3)
```

```python
df3 = df.merge(df2, how='inner', left_index = True, right_index= True)  # merging 1984-2008, with 2012-2016

df3.head(3)
```

```python
df2020 = pd.read_excel('pop-decennial.xls', sheet_name = 'States')
```

```python
df2020.set_index('Areaname', inplace = True)
```

```python
df2020 = df2020[[2020]]
```

```python
df2020.drop(index=['District of Columbia'], inplace = True)
```

```python
df_final = df3.merge(df2020, how='inner', left_index = True, right_index= True)  # merging 1984-2016, with 2020
```

```python
df_final.reset_index(inplace = True)
```

```python
df_final.rename(columns={'index': 'State'}, inplace = True)
```

# Get Turnout percentages

## 1980 - 2014

```python
# read files 

turnout_1980_2014 = pd.read_excel("1980-2014 November General Election.xlsx")
turnout_1980_2014
```

```python
# isolate columns needed
turnout_1980_2014 = turnout_1980_2014[["year", "state", "VEP Highest Office", "Highest Office", "Voting-Eligible Population (VEP)"]]

turnout_1980_2014 = turnout_1980_2014.loc[turnout_1980_2014["year"].isin(list(dataset_with_pop_incumbent_house_senate_spending.year.unique())), :]


display(turnout_1980_2014.head())

## convert to string to clean the currency signs
turnout_1980_2014 = turnout_1980_2014.astype("object")

## convert back to numbers
for col in ["VEP Highest Office", "Highest Office", "Voting-Eligible Population (VEP)"]:
    turnout_1980_2014[col] = turnout_1980_2014[col].astype("float64")
    
```

```python
# turnout
turnout_1980_2014
turnout_1980_2014.info()
```

## 2016 - 2020

```python
# read files
turnout_2016 = pd.read_excel("2016 - 2020 Turnout.xlsx", sheet_name = "2016")
turnout_2020 = pd.read_excel("2016 - 2020 Turnout.xlsx", sheet_name = "2020")
```

```python
# remove stars from names
turnout_2020["state"] = turnout_2020["state"].str.replace("*", "")
```

```python
# add years
turnout_2016["year"] = [2016 for i in range(turnout_2016.shape[0])]
turnout_2020["year"] = [2020 for i in range(turnout_2020.shape[0])]

display(turnout_2016)
display(turnout_2020)
```

## Concatenate

```python
# concat 2016 and 2020
turnout_2016_2020 = pd.concat([turnout_2016, turnout_2020], ignore_index=True)

turnout_2016_2018 = turnout_2016_2020.reindex(columns = turnout_1980_2014.columns)
turnout_2016_2020
```

```python
# concat 2016-2020 and 1980 - 2014

turnout = pd.concat([turnout_1980_2014, turnout_2016_2020])
display(turnout)
```

## Merge

```python
df_final = pd.merge(dataset_with_pop_incumbent_house_senate_spending, turnout, on = ["year", "state"], how = "left")
```

# Rearranging

```python
df_final.columns
```

```python
# rename columns
df_final.rename(columns = {"VEP Highest Office": "VEP_PCT_TURNOUT",
                          "Voting-Eligible Population (VEP)": "TOTAL_ELIGIBLE_POPULATION"}, inplace = True)
```

```python
# drop unneeded columns
df_final.drop("Highest Office", axis = 1, inplace = True)
```

```python
# sort it out by year
df_final = df_final.sort_values(by = "year")

# remove spaces
df_final.columns = df_final.columns.str.strip().str.replace(' ', '_')

# change it to one case
df_final.columns = df_final.columns.str.upper()

# view
display(df_final.head())

display(df_final.info())
```

# Clean datatypes


```python
# Clean

## convert to numeric
numerics = ["DEM_VOTES", "TOTALVOTES", "REP_VOTES", "THIRD_PARTY_VOTES", "MEDIAN_INCOME", "STANDARD_ERROR", "POPULATION", "NETVOTES"]

for col in numerics:
    df_final[col] = df_final[col].astype("float")
    
## convert rest to category
for col in df_final.select_dtypes("object"):
    df_final[col] = df_final[col].astype("category")
```

```python
# final view
display(df_final.info())
    
```
