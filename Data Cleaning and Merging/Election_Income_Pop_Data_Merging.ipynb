```python
import pandas as pd
import numpy as np
```

```python
income=pd.read_excel('h08.xlsx',sheet_name='2020_Dollars')
income
```

```python
income.columns[1:]
```

```python
z=[]
for x in range (2020,1983,-1):
    z.append('Median income_'+str(x))
    z.append('Standard error_'+str(x))
z
col_dict={list(income.columns[1:])[i]: z[i] for i in range(len(z))}
col_dict
```

```python
income=income.rename(columns=col_dict)
income=income.rename(columns={'Unnamed: 0': 'State'})
income
```

```python
election=pd.read_csv('1976-2020-president.csv')
election
```

```python
#election.to_excel('president.xlsx')
```

```python
election['party_detailed'].unique()
```

```python
election.party_simplified.unique()
```

```python
election_dem=election[election.party_simplified=='DEMOCRAT']
election_dem['Dem_candidate']=election_dem['candidate']
election_dem['Dem_votes']=election_dem['candidatevotes']
election_dem=election_dem[['year','state','Dem_candidate','Dem_votes','totalvotes']]

election_dem
```

```python
election_rep=election[election.party_simplified=='REPUBLICAN']
election_rep['Rep_candidate']=election_rep['candidate']
election_rep['Rep_votes']=election_rep['candidatevotes']
election_rep=election_rep[['year','state','Rep_candidate','Rep_votes','totalvotes']]

election_rep
```

```python
election_other=election[election.party_simplified=='OTHER']
election_lib=election[election.party_simplified=='LIBERTARIAN']
election_other=pd.concat([election_other,election_lib])
election_other=election_other.groupby(['year','state']).agg({'candidatevotes':'sum'})
election_other['Third_party_votes']=election_other['candidatevotes']
election_other=election_other.drop(columns=['candidatevotes'])
election_other
```

```python
election_other
```

```python
election=election_dem.merge(election_rep,how='inner',on=['year','state']).merge(election_other,how='inner',on=['year','state'])
election
```

```python
population=pd.read_excel('population.xlsx',sheet_name='1980-2020')
population
```

```python
population=population.rename(columns={'2020 Population':'Population_2020','2010 Population':'Population_2010','2000 Population':'Population_2000','1990 Population':'Population_1990','1980 Population':'Population_1980'})
```

```python
population
```

```python
population_long=pd.wide_to_long(population,stubnames='Population',i='State',j='Year',sep='_')
```

```python
population_long
```

```python
population_long.reset_index(level=0,inplace=True)
population_long.reset_index(level=0,inplace=True)
```

```python
population_long=population_long.rename(columns={'Year':'year','State':'state'})
```

```python
population_long
```
